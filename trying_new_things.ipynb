```python
from scipy import spatial

def strings_ranked_by_relatedness(
    query: str,
    df: pd.DataFrame,
    relatedness_fn=lambda x, y: 1 - spatial.distance.cosine(x, y),
    top_n: int = 10
) -> tuple[list[str], list[float]]:
    """
    Retorna as linhas do dataframe mais semelhantes a string de busca
    """

    query_embedding = text_embedding(query)

    strings_and_relatednesses = [ ]
    for i, row in df.iterrows():

        #relatedness = relatedness_fn(np.array(query_embedding).flatten(), np.array(row["embedding"]).flatten())
        relatedness = relatedness_fn(query_embedding, row["embeddings"])
        item = (row["texto_para_prompt"], relatedness)
        strings_and_relatednesses.append(item)

    strings_and_relatednesses.sort(key=lambda x: x[1], reverse=True)
    strings, relatednesses = zip(*strings_and_relatednesses)
    return strings[:top_n], relatednesses[:top_n]
```

```python
strings, relatednesses = strings_ranked_by_relatedness("Existe algum tipo de reclamação comum?",
                              df_menor)
for string, relatedness in zip(strings, relatednesses):
    print(f"{relatedness=:.3f}")
    display(string)
```

```python
import tiktoken

def num_tokens(text: str) -> int:
    encoding = tiktoken.encoding_for_model("gpt-3.5-turbo-16k")
    return len(encoding.encode(text))
```

```python
def montar_query(query: str, df: pd.DataFrame, token_budget: int) -> str:
    linhas, relatednesses = strings_ranked_by_relatedness(query, df)
    introducao = """
    Use o conteúdo abaixo com informações da ANEEL para responder a pergunta a seguir.
    As informações fornecidas devem ser sua única fonte para as respostas.
    Se não conseguir encontrar a resposta, escreva "Não consegui encontrar uma resposta."
    """
    pergunta = f"\n\Pergunta: {query}"
    mensagem = introducao
    for linha in linhas:
        proxima_linha = f'\n\nSeção com dados da ANEEL:\n"""\n{linha}\n"""'
        if num_tokens(mensagem + proxima_linha + pergunta) > token_budget:
            break
        else:
            mensagem = mensagem + proxima_linha

    return mensagem + pergunta
```

```python
def perguntar(
    query: str,
    df: pd.DataFrame = df,
    model: str = "gpt-35-turbo-16k",
    print_message: bool = False,
) -> str:
    message = montar_query(query, df, token_budget=16000)
    if print_message:
        print(message)
    messages = [
{"role": "system", "content": "Você está interagindo com o sistema especializado da ANEEL. Este sistema possui conhecimento abrangente sobre regulamentações, normas e operações no setor de energia elétrica. Sinta-se à vontade para fazer perguntas relacionadas a tarifas, regulações, qualidade dos serviços e outros aspectos do setor elétrico brasileiro. Forneça informações claras para obter respostas mais precisas e úteis."},
        {"role": "user", "content": message},
    ]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
        engine="gpt-35-turbo-16k"
    )
    response_message = response["choices"][0]["message"]["content"]
    return response_message
```
