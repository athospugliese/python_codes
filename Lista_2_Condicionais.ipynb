**Questão 5 - Ler**:

Você descobriu um sebo perto da sua casa e resolve dar uma olhada e encontra muitos títulos interessantes. Mas como você não está nadando em dinheiro, terá que fazer algumas escolhas. Escreva um programa que retorne se você deverá comprar ou não o livro, considerando que ele precisa estar na sua wishlist (nada de comprar por impulso) e o preço do sebo deve ser mais barato do que na internet.

a) Você deverá comprar Moby Dick?

b) Você deverá comprar Crime e Castigo?

c) Você deverá comprar a Odisseia?

````python
# RESPOSTA AQUI
wishlist = {
    "Moby Dick": {"sebo_price": 20, "online_price": 25},
    "Crime e Castigo": {"sebo_price": 35, "online_price": 20},
    "A Odisseia": {"sebo_price": 18, "online_price": 22}
}

def check_purchase(book):
    if book in wishlist:
        sebo_price = wishlist[book]["sebo_price"]
        online_price = wishlist[book]["online_price"]
        if sebo_price < online_price:
            return f"Você deve comprar o {book}, o preço no sebo é mais barato."
        else:
            return f"Não compre o {book}, o livro online é mais barato."
    else:
        return f"{book} não está na sua wishlist."

print(check_purchase("Moby Dick"))
print(check_purchase("Crime e Castigo"))
print(check_purchase("A Odisseia"))
print(check_purchase("Percy Jackson e os Olimpianos"))
````

**Questão 6 - Mercado futuro de festa**

Que as festas da faculdade são patrimônio universitário, todo mundo sabe, mas se a gente for em todas, não há carteira que aguente. Pior é pagar por uma festa agora e não ter grana para uma festa melhor depois. Assim, para lidar com essa volatilidade, você decide aplicar umas jogadas do mercado financeiro e de quebra, escrever uns códigos

Você paga $40 reais para seu amigo. Em troca, ele deverá pagar o seu ingresso para a próxima festa, independente do valor. Se o ingresso for mais caro, você se deu bem, se for mais barato, você se deu mal. Escreva um código que receba o input do valor do próximo ingresso, indique se foi uma operação vantajosa e o quanto você perdeu ou economizou.

````python
# RESPOSTA AQUI
valor_pago = 40

try:
    valor_ingresso_proximo = float(input("Insira o valor do próximo ingresso: "))

    if valor_ingresso_proximo >= valor_pago:
        economia = valor_ingresso_proximo - valor_pago
        print(f"Economia de ${economia:.2f}.")
    else:
        prejuizo = valor_pago - valor_ingresso_proximo
        print(f"Não economizou em ${prejuizo:.2f}.")
except ValueError:
    print("Digite um valor válido.")
````

**Questão 7 - Café com pão, café com pão...**

Você foi no Sweden na FEA tomar um café e saiu 254,76 reais. Na hora de efetuar o pagamento, você se deparou com algumas escolhas:

*   Você pode pagar a vista(1x)
*   Pode parcelar em até 3x com juros de 2,5% (aplicado sobre o valor inicial de 254,76)
*   Pode parcelar em mais vezes (máximo de 12 parcelas), com juros de 5%

Então faça um programa que calcule o valor total, dependendo da sua escolha, e o valor de cada parcela, caso opte por pagar a prazo.

````python
# RESPOSTA AQUI
def calcular_pagamento(valor, parcelas, juros):
    valor_total = valor * (1 + juros/100)
    valor_parcela = valor_total / parcelas
    return valor_total, valor_parcela

valor_compra = 254.76

opcao = int(input("Escolha a forma de pagamento:\n1 - À vista (1x)\n2 - Parcelado em até 3x com juros de 2,5%\n3 - Parcelado em mais vezes com juros de 5%\n"))

if opcao == 1:
    valor_total, _ = calcular_pagamento(valor_compra, 1, 0)
elif opcao == 2:
    valor_total, valor_parcela = calcular_pagamento(valor_compra, 3, 2.5)
elif opcao == 3:
    num_parcelas = int(input("Digite o número de parcelas (máximo 12): "))
    if num_parcelas <= 12:
        valor_total, valor_parcela = calcular_pagamento(valor_compra, num_parcelas, 5)

if opcao in [1, 2, 3]:
    print(f"Valor total a ser pago: R${valor_total:.2f}")
    if opcao in [2, 3]:
        print(f"Valor de cada parcela: R${valor_parcela:.2f}")
else:
    print("Opção de pagamento inválida.")
````
