```python
import pandas as pd
df = pd.read_csv("https://gist.githubusercontent.com/dmesquita/94d39a1291b5129fdecc51ec13e07b8d/raw/a4cdac97e8d94df4bbcf7d6cf2fe569e073429b5/df_reviews_livros.csv")

df.head(10)[["review_text"]]
```

```python
df.columns
```

```python
reviews = df["review_text"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
keywords = ['entrega', 'pacote', 'amassado', 'rasgado']

vectorizer = CountVectorizer(vocabulary=keywords)
```

```python
vector = vectorizer.transform(reviews)
```

```python
vector
```

```python
word_counts_df = pd.DataFrame(vector.toarray(), columns=keywords)
word_counts_df
```

```python
df['review_sobre_entrega'] = word_counts_df.sum(axis=1) > 0
```

```python
df['review_sobre_entrega'].value_counts(normalize=True)
```

```python
df.query("review_sobre_entrega == True")[["review_text"]]
```
